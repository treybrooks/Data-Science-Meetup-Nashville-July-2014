```python
import numpy as np
import pandas as pd
# http://developingcatan.files.wordpress.com/2011/02/settlers-of-catan-analysis.pdf
```

```python
players = ['Albert Camus','Jean-Paul Sartre', 'Franz Kafka']

weeks_in_a_year = 52

game_lengths = np.random.randint(50,100, weeks_in_a_year)
periods = pd.DatetimeIndex(start='7/7/2014', periods=weeks_in_a_year, freq='W').to_pydatetime()

games = zip(game_lengths, periods, range(len(periods)))

df = pd.DataFrame()
for game in games:
    g = pd.DataFrame({
        'Guildenstern': np.random.randint(1,7, game[0]), # {1,2,3,4,5,6}
        'Rosencrantz': np.random.randint(1,7, game[0]), # {1,2,3,4,5,6}
        'Period': game[1],
        'GameNumber': game[2],
        'Game_Length': game[0]
    })
    g['Roll'] = g.Guildenstern + g.Rosencrantz
    g = g.join(pd.DataFrame(players * (len(g)/len(players)+1), columns=['Player']))
    df = df.append(g)

# reorder columns
df = df[['Player', 'Period', 'GameNumber', 'Game_Length', 'Rosencrantz', 'Guildenstern', 'Roll']]
df['Progress'] = df.index
df.index = range(len(df))
print df.ix[len(df)-1]
```

#### Grouped DataFrame

```python
gdf = df.groupby(['Player','Period','GameNumber','Game_Length'], as_index=False).agg(mean)
gdf.index = gdf.Period
del gdf['Period'] # it was redundant
gdf = gdf.sort_index()
gdf.head()
```

## Python-NVD3

```python
import nvd3
from nvd3 import scatterChart
from nvd3 import stackedAreaChart
from nvd3 import lineWithFocusChart
```

### Scatter Chart

```python
#Open File to write the D3 Graph
output_file = open('ScatterChart_Long.html', 'w')

nicks = dict(zip(['camus','sartre','kafka'],players))

chart = scatterChart(name='scatterChart', height=700, width=600)

camus_r  = gdf[gdf.Player == nicks['camus']].Rosencrantz.map('{:,.2f}'.format).tolist()
sartre_r = gdf[gdf.Player == nicks['sartre']].Rosencrantz.map('{:,.2f}'.format).tolist()
kafka_r  = gdf[gdf.Player == nicks['kafka']].Rosencrantz.map('{:,.2f}'.format).tolist()
camus_g  = gdf[gdf.Player == nicks['camus']].Guildenstern.map('{:,.2f}'.format).tolist()
sartre_g = gdf[gdf.Player == nicks['sartre']].Guildenstern.map('{:,.2f}'.format).tolist()
kafka_g  = gdf[gdf.Player == nicks['kafka']].Guildenstern.map('{:,.2f}'.format).tolist()

kwargsC = {'shape': 'circle', 'size': '1'}
kwargsS = {'shape': 'cross', 'size': '1'}
kwargsK = {'shape': 'triangle-up', 'size': '1'}

extra_serie = {"tooltip": {"y_start": "", "y_end": "G", "y_format": "int", "x_start": "", "x_end": "R"}}

chart.add_serie(name=nicks['camus'], y=camus_g, x=camus_r, extra=extra_serie, **kwargsC)
chart.add_serie(name=nicks['sartre'], y=sartre_g, x=sartre_r, extra=extra_serie, **kwargsS)
chart.add_serie(name=nicks['kafka'], y=kafka_g, x=kafka_r, extra=extra_serie, **kwargsK)

#formating
chart.axislist['xAxis']['axisLabel'] = "'Rosencrantz'"
chart.axislist['yAxis']['axisLabel'] = "'Guildenstern'"

chart.margin_bottom = 40 #  can't see x label without it
chart.axislist['xAxis']['axisLabelDistance'] = 80
chart.axislist['yAxis']['axisLabelDistance'] = 40

chart.buildhtml()

output_file.write(chart.htmlcontent)

#close Html file
output_file.close()
```

### Stacked Area Chart

```python
output_file = open('stackedAreaChart_Long.html', 'w')

chart = stackedAreaChart(name='stackedAreaChart', height=400, width=800)

datax  = gdf.GameNumber.unique().astype(int).tolist()

camus  = gdf[gdf.Player == nicks['camus']].Roll.map('{:,.2f}'.format).astype(float).tolist()
sartre = gdf[gdf.Player == nicks['sartre']].Roll.map('{:,.2f}'.format).astype(float).tolist()
kafka  = gdf[gdf.Player == nicks['kafka']].Roll.map('{:,.2f}'.format).astype(float).tolist()

extra_serie = {"tooltip": {"y_start": "", "y_end": " ext"}}

chart.add_serie(name=nicks['camus'],  y=camus,  x=datax, extra=extra_serie)
chart.add_serie(name=nicks['sartre'], y=sartre, x=datax, extra=extra_serie)
chart.add_serie(name=nicks['kafka'],  y=kafka,  x=datax, extra=extra_serie)
chart.buildhtml()

output_file.write(chart.htmlcontent)

#close Html file
output_file.close()
```
