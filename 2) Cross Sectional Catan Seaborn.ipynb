### Get our data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

catan_df = pd.read_pickle('catan.pkl')
vc = catan_df.Roll.value_counts()
```

```python
plt.bar(vc.index, vc.values, align="center")
plt.show()
```

```python
import seaborn as sns
plt.bar(vc.index, vc.values, align="center")
plt.show()
```

```python
# style
sns.set(style="white", palette="muted")
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
sns.despine(left=True)
b, g, r, p = sns.color_palette("muted", 4)


# data
d = catan_df.Roll - 7

sns.distplot(d,             kde=False,           color=b,                          ax=axes[0, 0]) #blue
sns.distplot(d, hist=False,            rug=True, color=r,                          ax=axes[0, 1]) #red
sns.distplot(d, hist=False,                      color=g, kde_kws={"shade": True}, ax=axes[1, 0]) #green
sns.distplot(d,                                  color=p,                          ax=axes[1, 1]) #purple

plt.setp(axes, yticks=[])
plt.tight_layout()
```

### Joint Plots (2D Plots)

```python
sns.jointplot('Rosencrantz', 'Guildenstern', data=catan_df, kind="kde")
```

```python
sns.violinplot(catan_df, color="pastel")
```

```python
# data transform
catan_unstacked = catan_df[['Rosencrantz','Guildenstern']].unstack().reset_index().rename(columns={'level_0':'Dice','level_1':'Roll_Index', 0:'Roll',})
total_rolls = pd.DataFrame(catan_df.Roll)

#Clean up & Complicate 
silly_catan = pd.merge(catan_unstacked, total_rolls, left_on='Roll_Index', right_index=True)
silly_catan = silly_catan.reset_index()
silly_catan = silly_catan[['Dice','Roll_Index','Roll_x', 'Roll_y']]
silly_catan.columns = ['Dice','Roll_Index','Value', 'Roll_Sum']
silly_catan.head()
```
