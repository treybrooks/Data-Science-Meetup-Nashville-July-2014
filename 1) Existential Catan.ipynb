# CATAN!

#### "A weaker man might be moved to re-examine his faith, if in nothing else at least in the law of probability." - Guildenstern

#### Bar charts, histograms, etc.

```python
random.seed(8795)
dice_roll = lambda sides: random.randint(1,sides)
trails = 57
catan_game = [(dice_roll(6),dice_roll(6)) for _ in range(trails)]
catan_df = pd.DataFrame(catan_game, columns=['Rosencrantz', 'Guildenstern'])

catan_df['Roll'] = catan_df.Rosencrantz+catan_df.Guildenstern
catan_df.to_pickle('catan.pkl')
catan_df.head()
```

### Bar Charts

```python
#visualize the distribution from Pandas
vc = catan_df.Roll.value_counts(sort=False)

plt.close("all")

fig, ax = plt.subplots(1,1)
fig.set_facecolor('#589900')
fig.set_edgecolor('#589900')
ax.set_axis_bgcolor('#57046E')
plt.bar(vc.index, vc.values, align="center", color='#8D003E', alpha=1)
plt.axis("tight")
plt.xticks(range(2,13))
plt.xlabel("Roll")
plt.ylabel("Occurance")
plt.title("What if rolled some dice?")
plt.legend(["Roll"])
plt.grid()
plt.show()
plt.close("all")
```

### Pie Chart

```python
x = catan_df.Roll.copy()
pie = plt.pie(x)
plt.show(pie)
```

```python
import matplotlib.cm as cm
my_cmap = cm.rainbow(np.linspace(0, 1, 12))

pie = plt.pie(vc.values, labels=vc.index, colors=my_cmap)
plt.title("Distribution of rolls")
plt.xlabel("does this work")
plt.ylabel("it does!")

plt.show(pie)
```

### Scatter Plots

```python
j = 0.15
jitter = lambda lst: [ele + random.uniform(-j, j) for ele in lst]
plt.scatter(jitter(catan_df.Rosencrantz), jitter(catan_df.Guildenstern), c=catan_df.Roll, marker='o', cmap='hot')
plt.grid()
plt.xlabel('Rosencrantz')
plt.ylabel('Guildenstern')
```

### Subplots

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2)

ax1.bar(vc.index, vc.values, align="center", alpha=.5)
ax1.legend(["Roll"])

ax2.scatter(jitter(catan_df.Rosencrantz), jitter(catan_df.Guildenstern), c=catan_df.Roll, marker='o', cmap='hot')

ax3.hist(catan_df.Rosencrantz, np.linspace(1,7,7), normed=True, color="#6495ED", alpha=.5, align='left')
ax3.hist(catan_df.Guildenstern, np.linspace(1,7,7), normed=True, color="#F08080", alpha=.5, align='left')
ax3.legend(['R','G'], loc='best')

ax4.pie(vc.values, labels=vc.index, colors=my_cmap)

plt.show(pie)
```
